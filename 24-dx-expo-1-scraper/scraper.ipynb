```python
import requests
import bs4
url = 'https://app.sparkmailapp.com/web-share/scBxHGv__qleryQxHsRmgM98IG17-T3X-ZlQ3aQU'
page = requests.get(url)
soup = bs4.BeautifulSoup(page.content, 'html.parser')
```

```python
total_acc = []
```

```python


for i, e in enumerate(soup.find_all(id="exhibitor")):
    total_acc.append({"company": e.text.strip()})
```

```python
for i, ex in enumerate(soup.find_all(id="exhibitorHeadline")):
    total_acc[i]["headline"]=ex.text.strip()
```

```python
for i, ex in enumerate(soup.find_all(id="exhibitorDescription")):
    total_acc[i]["exhibitorDescription"]=ex.text.strip()
```

```python



connectionMadeBys = soup.find_all(id="connectionMadeBy")
acc = []
for i, connectionMadeBy in enumerate(connectionMadeBys):
    res = {}
    res["connectionMadeBy"] = connectionMadeBy.text
    ps = connectionMadeBy.find_next_siblings("p")
    for p in ps:
        a = p.find("a")
        if a["id"] == "exhibitorEmail":
            res["exhibitorEmail"] = p.text.strip()
        if a["id"] == "exhibitorPhone":
            res["exhibitorPhone"] = p.text.strip()
        if a["id"] == "exhibitorUrl":
            res["exhibitorUrl"] = p.text.strip()
    total_acc[i].update(res)
```

```python
import pandas as pd
df = pd.DataFrame(total_acc)
df.fillna("", inplace=True)
```
